```python
import pandas as pd
import numpy as np
import csv
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import re
from ipynb.fs.full.prepare_dataset import *
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df.head()
```

```python
p = df['Purpose'].unique()
purpose = {k:v for v,k in enumerate(p)}
```

```python
purpose
```

```python
for i in range(len(df)):
    df.loc[i,'Purpose'] = purpose[df.loc[i,'Purpose']]
```

```python
df.head()
df.drop(columns=['plane'],inplace=True)
```

```python
df.head()
```

```python
train,test = train_test_split(df,test_size=0.15)
train.reset_index(drop=True)
test.reset_index(drop=True)
```

```python
train_y = np.array(train['Credit amount'].tolist())
test_y = np.array(test['Credit amount'].tolist())
train_x = train.drop(columns=['Credit amount'])
test_x = test.drop(columns=['Credit amount'])
train_x.reset_index(drop=True,inplace=True)
test_x.reset_index(drop=True,inplace=True)
```

```python
TRAIN_x = []
for i in range(len(train_x)):
    TRAIN_x.append(train_x.loc[i,:])
TRAIN_x = np.array(TRAIN_x)
TEST_x = []
for i in range(len(test_x)):
    TEST_x.append(test_x.loc[i,:])
TEST_x = np.array(TEST_x)
```

```python
len(TRAIN_x)
```

```python
reg = LinearRegression().fit(TRAIN_x, train_y)
```

```python
Y = reg.predict(TEST_x)
```

```python
print('Coefficients: \n', reg.coef_)
```

```python
print('Mean squared error: %.2f'
      % mean_squared_error(test_y, Y))
# The coefficient of determination: 1 is perfect prediction
print('Coefficient of determination: %.2f'
      % r2_score(test_y, Y))
```
